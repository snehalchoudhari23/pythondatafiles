```python
import requests as rq
from bs4 import BeautifulSoup

url = 'https://quotes.toscrape.com/'
```

```python
resp = rq.get(url)

resp.content
```

```python
soupresp = BeautifulSoup(resp.content,'html.parser')
```

```python
print(soupresp.prettify())
```

```python
urlresp = soupresp.findAll('span', class_= 'text')
```

```python
for a in urlresp:
    print(a.text)
```

```python
b = soupresp.findAll('small', class_='author')
```

```python
for i in b:
    print(i.text)
```
